```python
import numpy as np
import pandas as pd
import tensorflow as tf
```

```python
tf.__version__
```

```python
dataset = pd.read_csv("Social_Network_Ads.csv")
x = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

```python
print(x)
```

```python
print(y)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
ann = tf.keras.models.Sequential()
```

```python
ann.add(tf.keras.layers.Dense(units = 6, activation = 'relu'))      #first layer
ann.add(tf.keras.layers.Dense(units = 6, activation = 'relu'))      #second layer
ann.add(tf.keras.layers.Dense(units = 1, activation = 'sigmoid'))      #output layer
```

```python
ann.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
ann.fit(x_train, y_train, batch_size=32, epochs = 100)
```

```python
# PREDICTING IF A PERSON AGED 30 WITH SALARY 135000 WILL PURCHASE
# PREDICTING IF A PERSON AGED 21 WITH SALARY 16000 WILL PURCHASE

print(ann.predict(sc.transform([[30, 135000]])) > 0.5)
print(ann.predict(sc.transform([[21, 16000]])) > 0.5)
```

```python
y_pred = ann.predict(x_test)
y_pred = (y_pred > 0)
print(np.concatenate((y_pred.reshape(len(y_pred), 1), y_test.reshape(len(y_test), 1)), 1))
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```
